```python
import torch
from diffusers import StableDiffusionPipeline

# make sure you're logged in with `huggingface-cli login`
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", revision="fp16", torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

# Image Generation and save to Google Drive


```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
def save_images(imgs, loc, text):
  for i, img in enumerate(imgs):
      img.save(f"{loc}/{text}_{i}.png")

def make_images_one_by_one(prompt, loc, filename, iter, sindex):
  for i in range(iter):
    image = pipe(prompt).images[0]
    image.save("%s/%s_%03d.png"%(loc, filename, sindex+i))

```

```python
location = './drive/MyDrive/stablediffusion/dataset'

startindex = 0
num_images = 50
prompt = ["an airplain flying in the sky"]
filename = 'airplane'

make_images_one_by_one(prompt, location, filename, num_images, startindex)
```
